This filtered dataset helps in analyzing how macroeconomic indicators may impact the *cement industry*.

```python
import pandas as pd

# Load the dataset
file_path = r"C:\Users\HP\Downloads\Worldbank_cleaned_2000_2023.csv"
df = pd.read_csv(file_path)

# Define keywords to filter the desired indicators
keywords = [
    "GDP", "Inflation", "Infrastructure", "Urban population",
    "Interest rate", "Construction", "CO2"
]

# Filter for rows where 'Series Name' contains any of the keywords
filtered_df = df[df["Series Name"].str.contains('|'.join(keywords), case=False, na=False)]

# Further filter for 'India' and 'World' only
filtered_df = filtered_df[filtered_df["Country Name"].isin(["India", "World"])]

# Save the filtered data to an Excel file
output_path = "Filtered_Indicators_India_World.xlsx"
filtered_df.to_excel(output_path, index=False)

print(f"Filtered data saved to {output_path}")
```

1.We read the Excel file containing various indicators and filter the columns corresponding to the years 2000 to 2023.                                                            2.We plot all the selected indicators on a single line chart to analyze trends and patterns over the years.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = r"C:\Users\HP\Desktop\New folder\Filtered_Indicators_India_World.xlsx"
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Extract year columns and years
year_columns = [col for col in df.columns if col.startswith('20')]
years = [int(col.split()[0]) for col in year_columns]

# Plotting all indicators
plt.figure(figsize=(14, 8))
for idx, row in df.iterrows():
    plt.plot(years, row[year_columns], label=row['Series Name'])

# Customize plot
plt.title("Economic Indicators (2000–2023)")
plt.xlabel("Year")
plt.ylabel("Value")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(True)
plt.tight_layout()
plt.show()
```

This section analyzes each macroeconomic indicator to detect the *largest year-over-year decline* from 2000 to 2023. These dips often highlight key economic shocks that may impact the cement industry.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = r"C:\Users\HP\Desktop\New folder\Filtered_Indicators_India_World.xlsx"
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Extract year columns and convert them
year_columns = [col for col in df.columns if col.startswith('20')]
years = [int(col.split()[0]) for col in year_columns]

# Loop through each indicator to detect dips
for idx, row in df.iterrows():
    values = row[year_columns].astype(float)
    changes = values.diff()

    # Find the year with the biggest drop
    min_change_col = changes.idxmin()              # column name like '2020 [YR2020]'
    min_change_year = int(min_change_col.split()[0])  # extract numeric year
    min_change_value = changes[min_change_col]
    
    print(f"{row['Series Name']}: Largest dip in {min_change_year} ({min_change_value:.2f})")

    # Plot with dip marked
    plt.figure(figsize=(10, 5))
    plt.plot(years, values, marker='o', label=row['Series Name'])
    plt.axvline(x=2008, color='red', linestyle='--', label='2008 Crisis')
    plt.axvline(x=2020, color='purple', linestyle='--', label='COVID-19')
    plt.axvline(x=min_change_year, color='black', linestyle=':', label=f'Biggest Dip: {min_change_year}')
    plt.title(f"{row['Series Name']} (2000–2023)")
    plt.xlabel("Year")
    plt.ylabel("Value")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
