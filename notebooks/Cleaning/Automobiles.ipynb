```python
import os
import pandas as pd
```

```python
folder_path = r"C:\Users\KARTHIKA\Kennedys[1]\Kennedys\Automobiles"
```

```python
# Correct sheet names from your files
sheet_names = ["balance_sheet", "cash_flow", "profit_&_loss", "quarters"]
```

```python
# Dictionary to store each cleaned sheet
cleaned_data = {}
```

```python
# Cleaning function
def clean_dataframe(df):
    df.dropna(how='all', inplace=True)                      # Drop empty rows
    df.dropna(axis=1, how='all', inplace=True)              # Drop empty columns
    df = df.loc[df.isnull().mean(axis=1) < 0.5]             # Remove rows >50% NaN
    df.columns = [str(col).strip() for col in df.columns]   # Clean column names
    df.fillna(method='ffill', inplace=True)                 # Fill missing values
    df.fillna(method='bfill', inplace=True)
     # Remove % signs and convert to numeric where applicable
    df = df.applymap(lambda x: str(x).replace('%', '') if isinstance(x, str) else x)
        
        # Convert all columns (except first one) to numeric if possible
    for col in df.columns[1:]:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    return df
```

```python
# Read and store each sheet separately
def process_excel_file(file_path):
    company_name = os.path.splitext(os.path.basename(file_path))[0]
    try:
        xls = pd.ExcelFile(file_path)
        for sheet in sheet_names:
            if sheet in xls.sheet_names:
                df = pd.read_excel(xls, sheet_name=sheet)
                df = clean_dataframe(df)
                cleaned_data[f"{company_name}_{sheet}"] = df
    except Exception as e:
        print(f"Error processing {file_path}: {e}")
```

```python
# Loop through files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.xlsx') or filename.endswith('.xls'):
        if filename.startswith('~$'):
            continue
        file_path = os.path.join(folder_path, filename)
        process_excel_file(file_path)
```

```python
def display_clean_summary(clean_dataframe):
    for sheet, df in cleaned_data.items():
        print(f"\nðŸ“„ Sheet: {sheet.title().replace('_', ' ')}")
        display(df.head(10))
```
